## REGRESSION MODEL

```python
y = data.price                   # variable to be explained
X = data.drop(columns="price")   # explanatory variables
```

```python
y.head()
```

```python
X.head()
```

```python
X.describe()
```

### Data Normalization 

```python
X.plot()
```

```python
X_norm = (X - X.mean()) / X.std() 
```

```python
X.describe()
```

```python
X.plot()
```

### Data Split : Train and Test

```python
import sklearn   
from sklearn.model_selection import train_test_split as separer   #on importe l'API qui permet de spliter
```

```python
X_train,X_test,y_train,y_test=separer(X_norm,y, test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

## Let's find out which model is best suited to our needs

```python
from lazypredict.Supervised import LazyRegressor

reg = LazyRegressor()

models,predictions = reg.fit(X_train, X_test, y_train, y_test)

print(models)
```

Apparently LGBMRegressor, XGBRegressor and HistGradientBoostingRegressor seem to be suitable for this regression problem. Let's try !

## LGBMRegressor

```python
from lightgbm import LGBMRegressor

# Initialize the model
lgbmr = LGBMRegressor()

# Train the model on the training data
lgbmr.fit(X_train, y_train)

# Evaluate the model performance
train_score = lgbmr.score(X_train, y_train)  # R² score on the training set
test_score = lgbmr.score(X_test, y_test)  # R² score on the test set

# Print the results
print(f"Train score: {train_score:.4f}")
print(f"Test score: {test_score:.4f}")
```

## XGBRegressor

```python
from xgboost import XGBRegressor  # Import the correct library

# Initialize the XGBoost Regressor
xgbr = XGBRegressor()

# Train the model on the training data
xgbr.fit(X_train, y_train)

# Evaluate the model performance
train_score = xgbr.score(X_train, y_train)  # R² score on the training set
test_score = xgbr.score(X_test, y_test)  # R² score on the test set

# Print the results
print(f"Train score: {train_score:.4f}")
print(f"Test score: {test_score:.4f}")
```

## HistGradientBoostingRegressor

```python
from sklearn.ensemble import HistGradientBoostingRegressor  # Correct import

# Initialize the Histogram-based Gradient Boosting Regressor
hgb_regressor = HistGradientBoostingRegressor()

# Train the model on the training data
hgb_regressor.fit(X_train, y_train)

# Evaluate the model performance
train_score = hgb_regressor.score(X_train, y_train)  # R² score on the training set
test_score = hgb_regressor.score(X_test, y_test)  # R² score on the test set

# Print the results
print(f"Train score: {train_score:.4f}")
print(f"Test score: {test_score:.4f}")
```

XGBRegressor seems to have the best results

## Error Analysis

```python
y_pred = xgbr.predict(X_test)
resultat = pd.DataFrame()
```

```python
resultat["reality"] = y_test
resultat["prediction"] = y_pred
resultat["error"] = abs(resultat["reality"]-resultat["prediction"])
resultat["error_relative"] = resultat["error"]/resultat["reality"]*100
resultat
```

```python
resultat.describe()
```

On average, we make errors of 13,13% in prices

```python
resultat.sort_values(by="error")
```

```python
resultat.error_relative.plot(kind="density")
```

```python
import scipy
scipy.stats.shapiro(resultat.error_relative)
```

Relative errors do not follow a normal distribution
